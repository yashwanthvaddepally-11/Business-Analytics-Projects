```python
import pandas as pd
df = pd.read_csv("D:/tiger analytics/Python/consumer behavior/ecommerce_customer_behavior_dataset_v2.csv") 
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
clv_df = df.groupby("Customer_ID").agg({
    "Total_Amount": "sum",      # Total revenue
    "Order_ID": "count",        # Number of orders
    "Date": ["min", "max"]      # Customer lifespan
})
```

```python
clv_df.head()
```

```python
clv_df.columns = ["Total_Revenue", "Total_Orders", "First_Order", "Last_Order"]
```

```python
clv_df["AOV"] = clv_df["Total_Revenue"] / clv_df["Total_Orders"]
```

```python
clv_df["Lifespan_Days"] = (clv_df["Last_Order"] - clv_df["First_Order"]).dt.days + 1
```

```python
clv_df.head()
```

```python
clv_df["CLV"] = clv_df["AOV"] * clv_df["Total_Orders"]
```

```python
clv_df.head()
```

```python
clv_df["Monthly_Frequency"] = clv_df["Total_Orders"] / (clv_df["Lifespan_Days"] / 30)

clv_df["Improved_CLV"] = clv_df["AOV"] * clv_df["Monthly_Frequency"] * 12  
```

```python
clv_df.head()
```

```python
clv_df.sort_values("Improved_CLV", ascending=False).head()
```

```python
clv_df.sort_values("Total_Revenue", ascending=False).head()
```

```python
clv_df.sort_values("AOV", ascending=False).head()
```

```python
behavior_df = df.groupby("Customer_ID").agg({
    "Session_Duration_Minutes": "mean",
    "Pages_Viewed": "mean",
    "Customer_Rating": "mean",
    "Is_Returning_Customer": "max"    # if returning at least once, consider as returning
})
behavior_df.columns = [
    "Avg_Session_Duration",
    "Avg_Pages_Viewed",
    "Avg_Rating",
    "Is_Returning_Customer"
]
```

```python
behavior_df.head()
```

```python
demo_df = df.groupby("Customer_ID").agg({
    "Age": "max",
    "Gender": "max",
    "City": "max"
})
```

```python
customer_df = clv_df.join(behavior_df).join(demo_df)
```

```python
customer_df.head()
```

```python
# Optional: encode categorical values
customer_df["Gender"] = customer_df["Gender"].astype("category").cat.codes
customer_df["City"] = customer_df["City"].astype("category").cat.codes
customer_df["Is_Returning_Customer"] = customer_df["Is_Returning_Customer"].astype("category").cat.codes
```

```python
print(customer_df.head())
print(customer_df.shape)
```

```python
customer_df = customer_df.drop(columns=["First_Order", "Last_Order"])
```
